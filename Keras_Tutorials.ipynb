```python
X = data.iloc[:, 1:22].values
y = data.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
#from sklearn.preprocessing import MinMaxScaler
#sc = MinMaxScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)
```

```python
inputs = keras.Input(shape=(21,), name="diabetes features")
layer_1 = layers.Dense(32, activation="relu", name="dense_1")(inputs)
layer_2 = layers.Dense(32, activation="relu", name="dense_2")(layer_1)
layer_3 = layers.Dense(32, activation="relu", name="dense_3")(layer_2)
layer_4 = layers.Dense(32, activation="relu", name="dense_4")(layer_3)
layer_5 = layers.Dense(32, activation="relu", name="dense_5")(layer_4)
layer_6 = layers.Dense(32, activation="relu", name="dense_6")(layer_5)
layer_7 = layers.Dense(32, activation="relu", name="dense_7")(layer_6)
layer_8 = layers.Dense(32, activation="relu", name="dense_8")(layer_7)
layer_9 = layers.Dense(32, activation="relu", name="dense_9")(layer_8)
layer_10 = layers.Dense(32, activation="relu", name="dense_10")(layer_9)
layer_11 = layers.Dense(32, activation="relu", name="dense_11")(layer_10)
layer_12 = layers.Dense(32, activation="relu", name="dense_12")(layer_11)
layer_13 = layers.Dense(32, activation="relu", name="dense_13")(layer_12)
layer_14 = layers.Dense(32, activation="relu", name="dense_14")(layer_13)
layer_15 = layers.Dense(32, activation="relu", name="dense_15")(layer_14)
layer_16 = layers.Dense(32, activation="relu", name="dense_16")(layer_15)
outputs = layers.Dense(1, activation="sigmoid", name="predictions")(layer_16)

model = keras.Model(inputs=inputs, outputs=outputs)
```

```python
model.compile(
    optimizer=keras.optimizers.SGD(),
    loss=keras.losses.binary_crossentropy,
    metrics=[keras.metrics.mean_squared_error]
)
```

```python
print("Fit the model on the training data")
history = model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=50
)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
```

```python
print("Evaluate the test data")
results = model.evaluate(X_test, y_test, batch_size=64)
print("test loss, test accuracy:", results)

print("Generate predictions for 3 samples")
predictions = model.predict(X_test[:3])
print("predictions shape:", predictions.shape)
print(predictions)
```
