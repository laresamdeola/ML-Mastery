```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X, y)
print("Original dataset", Counter(y))
print("Resample dataset shape", Counter(y_smote))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size=0.25, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=4, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=4, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=4, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy", "mse"])
```

```python
ann.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.78)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
```

```python
print(cm)
```

```python
print(acc)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("Specificity: ", tn)
print("Sensitivity: ", tp)
```

```python
from sklearn.metrics import classification_report
target = ["class 0", "class 1"]
print(classification_report(y_test, y_pred, target_names=target))
```
