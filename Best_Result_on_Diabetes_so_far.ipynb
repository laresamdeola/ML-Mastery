```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
high_blood_pressure = X_train[:, 0]
```

```python
print(high_blood_pressure)
```

```python
import keras

inputs = keras.Input(shape=(21,), name="Input Layer")
layer_1 = keras.layers.Dense(32, activation="relu", name="dense_layer_1")(inputs)
layer_2 = keras.layers.Dense(32, activation="relu", name="dense_layer_2")(layer_1)
outputs = keras.layers.Dense(3, activation="sigmoid", name="output_layer")(layer_2)
```

```python
model = keras.Model(inputs=inputs, outputs=outputs, name="Diabetes_Model")
```

```python
model.compile(optimizer=keras.optimizers.RMSprop(), loss=keras.losses.SparseCategoricalCrossentropy(), metrics=[keras.metrics.SparseCategoricalAccuracy()])
```

```python
model.summary()
```

```python
print("Fitting the model on the training data")
history = model.fit(X_train, y_train, batch_size=64, epochs=10)
```

```python
test_scores = model.evaluate(X_test, y_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
```

```python
predictions = model.predict(X_test)
print(predictions.shape, y_test.shape)
```
